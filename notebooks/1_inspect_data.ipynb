## Dataset characterization
Here are a few samples and statistics to help get a feel for the data.
We see that there are 150 unique texts equally distributed over the three splits.
Texts are short, usually between 5 and 20 words.
Label distribution is balanced.

```python
import pandas as pd

df = pd.concat(
    [
        pd.read_json(f"../data/swedish_sentiment_{split}.jsonl", lines=True).assign(split=split)
        for split in ["train", "val", "test"]
    ]
)

pd.set_option('display.max_colwidth', None)
df.sample(5)
```

```python
df.describe()
```

```python
df["length"] = df["text"].str.split().str.len()
df["positive"] = df["label"] == "positive"

df.groupby("split").agg({"text": "count", "length": ["mean", "min", "max"], "positive": "mean"})
```
